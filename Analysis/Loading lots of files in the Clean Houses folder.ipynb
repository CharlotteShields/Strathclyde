### Data analysis of the Strathclyde REFIT data

A notebook which demonstrates analysis of the Strathclyde dataset

#### Setup

```python
import pandas as pd
from pandas import *
import os
filepath=r'C:\Users\cvcs4\OneDrive - Loughborough University\Data Analysis\Refit\Clean House'
## r infront of '' means the string is raw - you need this if it's a 
## filepath because backspaces do funny things to the 
## string if you don't tell python it's raw.
## The above filepath is for the folder containing the files.
```

### Data processing

This loads the data, resamples it to half hour intervals and saves the raw data files.
We need to loop through files in a directory (a directory is a folder).

```python
flag=True
for filename in os.listdir(filepath):
    print(filename)
    path=os.path.join(filepath,filename)
    ## the above needs doing because the filepath is not in the same
    ## folder that we're working in here
    
    ## parse_dates=['Time'] is a better way of getting the time column to work
    df=pd.read_csv(path,parse_dates=['Time'])
    
    ## set index as time column
    df=df.set_index('Time')
    
    ## resample the data to work with half hour intervals
    ## T = minutes, M=monthly
    df=df.resample('30T').mean()
    
    ## remove time from index and set it to 0,1,2 etc
    df=df.reset_index()
    
    ## create a column showing the filename
    df['file']=filename
    
    ## This flag function makes it add each file's dataframe to the last (append)
    ## This will be useful! See the flag=True at the top of the loop
    if flag:
        df2=df
        flag=False
    else:
        df2=df2.append(df)
    
    ## For the sake of seeing if everything works, break breaks the loop.
    ## Femove this when you want to run the entire loop.
    # break
    
    ## Save the data extracted and appended (above) to a new massive file
    ## containing all the data from the files in the folder.
df2.to_csv('NewData.csv')
```
